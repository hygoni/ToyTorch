```python
import matplotlib.pyplot as plt
```

```python
from torch.autograd import Variable
import numpy as np
import torch

time = [1, 3, 5, 7, 10, 17, 18]
grade = [2.0, 2.5, 3.5, 3.0, 4.0, 4.0, 4.5]

plt.title('TIme vs Grade')
plt.xlabel('Time spent studying')
plt.ylabel('Grade')
plt.plot(time, grade, 'o')

time = np.array(time)
time = time.reshape(-1, 1)
time = Variable(torch.Tensor(time), requires_grad = True)

grade = np.array(grade)
grade = time.reshape(-1, 1)
grade = Variable(torch.Tensor(grade), requires_grad = True)

print(time, grade)
```

### Building model

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

class LinearRegression(nn.Module):
    def __init__(self, in_size, out_size):
        super(LinearRegression, self).__init__()
        self.linear = nn.Linear(in_size, out_size)
    def forward(self, x):
        return self.linear(x)
```

### Training

```python
in_size = 1
out_size = 1
net = LinearRegression(in_size, out_size)

optimizer = optim.Adam(net.parameters(), lr = 0.01)
get_loss = nn.MSELoss()

epoch_num = 300
loss_graph = []

for i in range(epoch_num):
    optimizer.zero_grad()
    out = net(time)
    loss = get_loss(out, grade)
    loss.backward()
    optimizer.step()
    loss_graph.append(loss.data)
    #print(loss.data)
```

```python
plt.plot(loss_graph)
```

```python
x = np.arange(1, time.detach().numpy()[-1])
x = x.reshape(-1, 1)
x = torch.Tensor(x)
y = net(x)
plt.plot(x.detach().numpy(), y.detach().numpy())
plt.plot(time.detach().numpy(), grade.detach().numpy(), 'o')

out = net(time)
print('Mean Squared Error : {}'.format(nn.MSELoss()(out, grade)))
```
